```python
size(401, 401)

w = 40
count = 0
num_cols = num_rows = 10
for col in range(num_cols):
    x = col * w
    for row in range(num_rows):
        y = row * w 
        fill(255)
        square(x, y, w)
        fill(0)
        text(count, x + w / 2, y + w / 2)
        count += 1
```

```python
size(401, 401)

w = 40
num_cols = num_rows = 10
for i in range(num_cols):
    x = i * w
    for j in range(num_rows):
        y = j * w 
        back, front = 0, 255
        if (i + j) % 2:
            front, back = back, front
        fill(back)
        square(x, y, w)
        fill(front)
        text(f'{i},{j}', x + w / 2, y + w / 2)
```

```python
size(401, 401)

from itertools import product

w = 40
count = 0
order = 10
rc = product(range(order), repeat=2)
for row, col in rc:
    x, y = col * w, row * w 
    fill(255)
    square(x, y, w)
    fill(0)
    text(count, x + w / 2, y + w / 2)
    count += 1
```

```python
size(401, 401)
w = 40
text_size(20)
text_align(CENTER, CENTER)
x = y = 0
c = 32
while c < 127:
    no_fill()
    square(x, y, w)
    text(chr(c), x + w / 2, y + w / 2)
    x += w
    if x > width - w:
        x = 0
        y += w
    c += 1
```
